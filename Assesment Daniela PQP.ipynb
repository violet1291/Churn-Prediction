# Creación de los clústers

```python
#Importando KMeans
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
#Definiendo el número óptimo de clústers
```

```python
#Definiendo el nivel óptimo de clústers
def optimise_k_means(data, max_k):
    means = []
    inertias = []

    for k in range(1, max_k):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        means.append(k)
        inertias.append(kmeans.inertia_)

    # Generate the elbow plot
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(means, inertias, 'o-')
    ax.set_xlabel('Number of Clusters', fontsize=12)
    ax.set_ylabel('Inertia', fontsize=12)
    ax.set_title('Elbow Method for Optimal k', fontsize=14)
    ax.grid(True)
    plt.show()
```

```python
optimise_k_means(df.drop(['Fuga','Var de la Prioridad del Soporte'], axis=1), 10)
```

```python
#El número de clústers óptimos es de aproximadamente 2-3 clústers
```

```python
#3 clústers
kmeans3 = KMeans(n_clusters=3)
#Se suele recomendar usar un número de clústers impares ya que se suelen ajustar mejor a la data
```

```python
kmeans3.fit(df.drop(['Fuga','Var de la Prioridad del Soporte'],axis=1))
```

```python
#Centroides
kmeans3.cluster_centers_
```

```python
#Agregando variable de clasificación de clústers al data frame
df['k_means3']= kmeans3.labels_
```

```python
df.head()
```

```python
#Visualización de clústers con PCA

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Datos usados para entrenar KMeans
X = df.drop(['Fuga','Var de la Prioridad del Soporte'], axis=1)

# Reducción de dimensionalidad a 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)

# Añadimos resultados al DataFrame
df_plot = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])
df_plot['Cluster'] = kmeans3.labels_

# Graficamos
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_plot, x='PC1', y='PC2', hue='Cluster', palette='Set2', s=60)
plt.title('Visualización de Clústers con PCA')
plt.show()
```

# Regresión Logística

```python
df.head()
```

```python
X= df.drop('Fuga', axis=1)
y= df['Fuga']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train, y_train)
predictions = logmodel.predict(X_test)
```

```python
#Reporte de clasificación del modelo logístico
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
#Ecuación logística

# Intercepto (β₀)
intercepto = logmodel.intercept_[0]

# Coeficientes (β₁, β₂, …)
coeficientes = logmodel.coef_[0]

# Nombres de las variables
variables = X.columns

# Mostrar la ecuación
print("Ecuación logística:")
ecuacion = f"logit(p) = {intercepto:.4f}"
for var, coef in zip(variables, coeficientes):
    signo = '+' if coef >= 0 else ''
    ecuacion += f" {signo}{coef:.4f}*{var}"
print(ecuacion)
```

```python
odds_ratios = np.exp(logmodel.coef_[0])
print(odds_ratios)
```

```python
#Matriz de variable por coeficiente y Odds ratio
resultado = pd.DataFrame({
    'Variable': variables,
    'Coeficiente (β)': coeficientes,
    'Odds Ratio (e^β)': np.exp(coeficientes),
    'Efecto': ['Aumenta la probabilidad de fuga' if c > 0 else 'Disminuye la probabilidad de fuga' for c in coeficientes]
})

# Redondear para presentación
resultado = resultado.round(4)
```

```python
#Exportando excel
resultado.to_excel('Resultados_Logit.xlsx', index=False)
```
